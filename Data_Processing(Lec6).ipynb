```python
import pandas as pd
exam=pd.read_csv('exam.csv')

exam.query('nclass==1') # 1반 학생들의 데이터만 추출 (df.query()는 행 추출)
exam.query('math>50') # 수학 점수가 50점 초과한 데이터만 추출
exam.query('nclass==1 & math>=50') # 여러 조건을 동시에 충족하는 행 추출
exam.query('math>=90 | english>=90') # 여러 조건 중 하나라도 충족하는 행 추출

exam.query('nclass == 1 | nclass==3 | nclass==5') # 밑의 코드와 동일
exam.query('nclass in [1,3,5]')
```

<h4>집단별로 요약하기</h4>

---



<전체 요약 통계랑 구하기>

```python
import pandas as pd
exam=pd.read_csv('exam.csv')

exam.agg(mean_math=('math','mean')) # 요약 시 사용할 변수명과 함수명은 따옴표로 감싸 문자형태로, 함수명 뒤에 () 안쓴다 !
```

<집단별 요약 통계량 구하기>

```python
# 반 별 수학 점수 평균 구하기
exam.groupby('nclass')\
    .agg(mean_math=('math','mean'))

# 변수를 인덱스로 바꾸지 않기 (groupby()의 기본값이 변수를 인덱스로 바꾸게 함 !)
exam.groupby('nclass',as_index=False)\
    .agg(mean_math=('math','mean'))
```

<여러 요약 통계량 한 번에 구하기>

```python
exam.groupby('nclass')\
    .agg(mean_math=('math','mean'),
         sum_math=('math','sum'),
         median_math=('math','median'),
         n=('nclass','count'))
```

<집단별로 다시 집단 나누기>

```python
mpg=pd.read_csv('mpg.csv')
mpg.groupby(['manufacturer','drv'])\
   .agg(mean_cty=('cty','mean'))
```

**Practice** : midwest.csv 데이터로 분석 문제 해결

1. midwest 데이터에 '전체 인구 대비 미성년 인구 백분율' 변수 추가
2. 미성년 인구 백분율이 가장 높은 상위 5개 county(지역)의 미성년 인구 백분율 출력
3. 분류표의 기준에 따라 미성년 비율 등급 변수 추가하고, 각 등급에 몇 개의 지역이 있는지 알아보기 (기준 : 40% 이상 -> large, 30~40% -> middle, 30% 이하 -> small)
4. '전체 인구 대비 아시아인 인구 백분율' 변수를 추가하고, 하위 10개 지역의 state, county, 아시아인 인구 백분율 출력

```python
midwest=pd.read_csv('midwest.csv')

# Problem 1
midwest=midwest.assign(not_adult_ratio=(1-midwest['popadults']/midwest['poptotal'])*100)
midwest.head()
```

```python
# Problem 2
midwest.sort_values('not_adult_ratio',ascending=False)[['county','not_adult_ratio']].head()
```

```python
# Problem 3
import numpy as np
midwest=midwest.assign(not_adult_class=np.where(midwest['not_adult_ratio']>=40,'large',\
                                                np.where(midwest['not_adult_ratio']>=30,'middle','small')))
midwest['not_adult_class'].value_counts()
```

```python
# Problem 4
midwest=midwest.assign(asian_ratio=midwest['popasian']/midwest['poptotal']*100)
midwest.sort_values('asian_ratio')[['state','county','asian_ratio']].head(10)
```
