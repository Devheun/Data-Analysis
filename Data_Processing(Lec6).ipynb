<h4>조건에 맞는 데이터만 추출하기<h4>

---


```python
import pandas as pd
exam=pd.read_csv('exam.csv')

exam.query('nclass==1') # 1반 학생들의 데이터만 추출 (df.query()는 행 추출)
exam.query('math>50') # 수학 점수가 50점 초과한 데이터만 추출
exam.query('nclass==1 & math>=50') # 여러 조건을 동시에 충족하는 행 추출
exam.query('math>=90 | english>=90') # 여러 조건 중 하나라도 충족하는 행 추출

exam.query('nclass == 1 | nclass==3 | nclass==5') # 밑의 코드와 동일
exam.query('nclass in [1,3,5]')
```

<추출한 행으로 데이터 만들기>

```python
nclass1=exam.query('nclass==1') # 새 변수에 할당하면 됨
nclass2=exam.query('nclass==2') 
nclass1
```

```python
df=pd.DataFrame({'sex':['F','M','F','M'],
                'country':['Korea','China','Japan','USA']})

df.query('sex=="F" & country=="Korea"') # 따옴표 다르게 입력해야함
```

**Practice** : mpg 데이터를 이용해 분석 문제 해결해보기
1. displ(배기량)이 4 이하인 자동차와 5 이상인 자동차 중 어떤 자동차의 hwy(고속도로 연비) 평균이 더 높은지 알아보기

2. 'audi'와 'toyota'중 어느 manufacturer(자동차 제조회사)의 cty(도시 연비) 평균이 더 높은지 알아보기

3. 'chevrolet', 'ford', 'honda' 자동차의 고속도로 연비 평균을 알아 볼 때, 세 회사의 데이터 추출 후 hwy 전체 평균 구해보기


```python
# Problem 1
mpg=pd.read_csv('mpg.csv')
displ_4=mpg.query('displ<=4') # 배기량이 4 이하인 자동차
displ_5=mpg.query('displ>=5') # 배기량이 5 이상인 자동차
displ_4['hwy'].mean(), displ_5['hwy'].mean() # 배기량이 4 이하인 자동차가 고속도로 연비 평균이 더 높다
```

```python
# Problem 2
audi=mpg.query('manufacturer=="audi"')
toyota=mpg.query('manufacturer=="toyota"')
audi['cty'].mean(), toyota['cty'].mean() # toyota 자동차가 도시 연비 평균이 더 높다
```

```python
# Problem 3
cfh=mpg.query('manufacturer in ["chevrolet","ford","honda"]')
cfh['hwy'].mean()
```

<h4>필요한 변수만 추출하기</h4>

---



```python
exam['math'] # df[]로 변수 추출 가능
exam[['nclass','math','english']] # 여러 변수 함께 추출

# 추가적으로 알아둘 것 : 변수를 1개만 추출하면 시리즈(series) 자료구조로 바뀌는데, 데이터프레임 자료구조를 유지하려면 []로 한번 더 감싼다.
# ex ) exam['math']는 시리즈, exam[['math']]는 데이터 프레임
```

<변수 제거하기>

```python
# df.drop()
exam.drop(columns='math') # math 변수 제거
exam.drop(columns=['math','english']) # 여러 변수 제거
```

<pandas 함수 조합하기>

```python
# query()와 df[] 조합하기

exam.query('nclass==1')['english'] # 1반 학생의 영어점수 추출
exam.query('math>=50')[['id','math']] # 수학 점수가 50점 이상인 학생의 id와 math 변수
exam.query('math>=50')[['id','math']].head() # 위의 추출결과에서 일부만 출력
```
